```python
def score_classifier(dataset,classifier,labels):

    """
    performs 3 random trainings/tests to build a confusion matrix and prints results with precision and recall scores
    :param dataset: the dataset to work on
    :param classifier: the classifier to use
    :param labels: the labels used for training and validation
    :return:
    """
    
  
    
    kf = KFold(n_splits=3,random_state=50,shuffle=True)
    confusion_mat = np.zeros((2,2))
    recall = 0
    y_true = []
    y_pred = []
    

    for training_ids,test_ids in kf.split(dataset):
        training_set = dataset[training_ids]
        training_labels = labels[training_ids]
        test_set = dataset[test_ids]
        test_labels = labels[test_ids]
        ##Select the Kbest features
        ##classifier=make_pipeline(PolynomialFeatures(2), SelectKBest(f_classif, k=5), RandomForestClassifier(random_state=50) )



        classifier.fit(training_set,training_labels)
        predicted_labels = classifier.predict(test_set)
        confusion_mat+=confusion_matrix(test_labels,predicted_labels)
        classification_report(test_labels,predicted_labels)
        recall += recall_score(test_labels, predicted_labels)
        y_true.append(test_labels)
        y_pred.append(predicted_labels)
        #print(y_true)
        #print(y_pred)
    N, train_score, val_score = learning_curve(classifier, training_set,training_labels, cv=3, scoring='recall', train_sizes=np.linspace(0.1,1,10))
    recall/=3
    
    y_true=np.concatenate(y_true)
    y_pred=np.concatenate(y_pred)
    precision, recall, threshold = precision_recall_curve(test_labels, classifier.best_estimator_.decision_function(test_set))
    plt.figure()
    
    plt.title('Precision Recall Curve')
    plt.plot(threshold,precision[:-1],label='precision')
    plt.plot(threshold,recall[:-1],label='recall')
    plt.legend()
    
    print(confusion_mat)
    #print(recall)
    classe=classification_report(y_true,y_pred)
    print(classe)
    
    plt.figure()
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='val score')
    #plt.title('K=5')
    plt.legend()
    print(classifier.best_params_)
    y_pred=classifier.predict(training_set)
    
    ###Feature slection the k best###
   
```
